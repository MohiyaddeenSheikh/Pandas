```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Database/Superstore-Sales.csv",encoding="latin1")
```

```python
pivot = df.pivot_table(
                index = 'Ship Mode',
                columns = ['Product Category'],
                values = 'Profit',
                aggfunc = 'sum'
                
)
pivot2 = df.pivot_table(index = 'Ship Mode', columns = ['Product Category'], values = 'Sales', aggfunc = 'sum')

# Plotting the heatmap
fig = plt.figure(figsize=(18, 4))
gs = fig.add_gridspec(1, 2)

sns.heatmap(
    ax=fig.add_subplot(gs[0, 0]),
    data=pivot,
    cmap='coolwarm',
    annot=True,
    fmt='.0f',
    cbar=True
)


sns.heatmap(
    ax=fig.add_subplot(gs[0, 1]),
    data=pivot2,
    cmap='coolwarm',
    annot=True,
    fmt='.0f',
    cbar=True
)


plt.show()
```
