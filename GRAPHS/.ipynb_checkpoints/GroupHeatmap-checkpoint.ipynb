```python
import pandas as pd
df = pd.read_csv('./Database/Superstore-Sales.csv',encoding ="latin1")

group_sales = df.groupby('Ship Mode')[['Sales']]
                .agg('sum')
                .sort_values(
                            by="Sales",
                            ascending = False
                            )

group_shippingcost = df.groupby('Ship Mode')[['Shipping Cost']].agg('sum').sort_values(by="Shipping Cost",ascending = False)
group_ProductCategory = df.groupby('Product Category')[['Profit']].agg('sum').sort_values(by="Profit",ascending = False)
group_productsubcategory = df.groupby('Product Sub-Category')[['Profit']].agg('sum').sort_values(by="Profit",ascending = False)


import seaborn as sns
import matplotlib.pyplot as plt

# plt => figure => add_gridspec => add_subplot => ax => heatmap
# Create a 2x3 grid using GridSpec
fig = plt.figure(figsize=(17, 10))
gs = fig.add_gridspec(2, 3)
 
sns.heatmap( 
            ax=fig.add_subplot(gs[0, 1]), 
            data = group_sales, 
            cmap= 'coolwarm' , 
            annot =True, 
            fmt = ".0f",
            cbar=True
            )
sns.heatmap(ax=fig.add_subplot(gs[0, 2]),data = group_shippingcost, cmap= 'coolwarm' , annot =True, fmt = ".0f",cbar=True)
sns.heatmap(ax=fig.add_subplot(gs[1, 1]),data = group_ProductCategory, cmap= 'coolwarm' , annot =True, fmt = ".0f",cbar=True)
sns.heatmap(ax=fig.add_subplot(gs[0:2, 0]) ,data = group_productsubcategory, cmap= 'coolwarm' , annot =True, fmt = ".0f",cbar=True)
```

# ðŸ“Š Creating a Multi-Grid Heatmap Layout in Matplotlib

**Workflow:**  
**plt** <span style="color:#e74c3c; font-weight:bold;">â†’</span> **figure** <span style="color:#2980b9; font-weight:bold;">â†’</span> **add_gridspec** <span style="color:#27ae60; font-weight:bold;">â†’</span> **add_subplot** <span style="color:#f39c12; font-weight:bold;">â†’</span> **ax** <span style="color:#8e44ad; font-weight:bold;">â†’</span> **heatmap**

---

### Steps

- **plt** <span style="color:#e74c3c; font-weight:bold;">â†’</span>  
  <span style="color:#e74c3c; font-weight:bold;">Create the base for plotting using Matplotlib's pyplot.</span>

- **figure** <span style="color:#2980b9; font-weight:bold;">â†’</span>  
  <span style="color:#2980b9; font-weight:bold;">Initialize a new figure to hold your plots.</span>

- **add_gridspec** <span style="color:#27ae60; font-weight:bold;">â†’</span>  
  <span style="color:#27ae60; font-weight:bold;">Define the grid layout (rows and columns) for flexible subplot arrangement.</span>

- **add_subplot** <span style="color:#f39c12; font-weight:bold;">â†’</span>  
  <span style="color:#f39c12; font-weight:bold;">Add a subplot to a specific location in the grid (can span multiple rows/columns).</span>

- **ax** <span style="color:#8e44ad; font-weight:bold;">â†’</span>  
  <span style="color:#8e44ad; font-weight:bold;">Reference to the axes object where you will plot your data.</span>

- **heatmap**  
  <span style="color:#16a085; font-weight:bold;">Plot your heatmap using Seaborn on the specified axes.</span>


# ðŸ“Š Workflow: Summing and Sorting Profit by Product Sub-Category

**df** <span style="color:#e74c3c; font-weight:bold;">â†’</span> **groupby('Product Sub-Category')** <span style="color:#2980b9; font-weight:bold;">â†’</span> **[['Profit']]** <span style="color:#27ae60; font-weight:bold;">â†’</span> **agg('sum')** <span style="color:#f39c12; font-weight:bold;">â†’</span> **sort_values(by="Profit", ascending=False)**

---

### Steps

- **df** <span style="color:#e74c3c; font-weight:bold;">â†’</span>  
  <span style="color:#e74c3c; font-weight:bold;">Start with your main DataFrame.</span>

- **groupby('Product Sub-Category')** <span style="color:#2980b9; font-weight:bold;">â†’</span>  
  <span style="color:#2980b9; font-weight:bold;">Group the data by each unique product sub-category.</span>

- **[['Profit']]** <span style="color:#27ae60; font-weight:bold;">â†’</span>  
  <span style="color:#27ae60; font-weight:bold;">Select only the 'Profit' column from each group for aggregation.</span>

- **agg('sum')** <span style="color:#f39c12; font-weight:bold;">â†’</span>  
  <span style="color:#f39c12; font-weight:bold;">Sum the profits for each product sub-category group.</span>

- **sort_values(by="Profit", ascending=False)**  
  <span style="color:#8e44ad; font-weight:bold;">Sort the results by total profit in descending order (highest profit first).</span>
