```python
import pandas as pd
df = pd.read_csv("./Database/Superstore-Sales.csv",encoding = "latin1")
numeric_col = df[['Sales','Discount','Profit','Shipping Cost','Unit Price']]
df_corr = numeric_col.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
fig = plt.figure(figsize=(18, 4))
gs = fig.add_gridspec(1,2)

sns.heatmap(
            ax = fig.add_subplot(gs[0,0]),
            data = df_corr,
            annot = True,
            fmt = '.3f',
            mask = np.triu(np.ones_like(df_corr, dtype=bool)) # remove upperpart
)
```
